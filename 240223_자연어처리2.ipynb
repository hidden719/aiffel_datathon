## 2. 해당 데이터의 중복된 문자열 제거 EX ) 영양소, 원료명, 제품타입, 섭취방법, 섭취대상: 이런것들? 

```python
df = info.copy()
```

```python
columns_to_clean = ['제품특성']
words_to_remove = ['제품타입','제품유형','영양소','식약처고시','제품용량','제품타입','원료명','하루','등급','구성','기능']
```

```python
for col in columns_to_clean:
    for word in words_to_remove:
        df[col] = df[col].str.replace(word, '')

print(df)
```

## 3. 거기서 이제 토큰화해서 유니크값 키워드 추출해보기 

```python
df['제품특성'].unique()
```

```python
import pandas as pd
from konlpy.tag import Okt
tokenizer = Okt()

text_series = df['제품특성'].astype(str)

tokens = text_series.apply(lambda x: tokenizer.morphs(x))
```

```python
tokens
```

```python
df['제품특성'].unique()
```

```python
nouns = text_series.apply(lambda x: tokenizer.nouns(x))
nouns
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()

token_string = [' '.join(token_list) for token_list in tokens]
vect.fit(token_string)
```

```python
vect.vocabulary_
```

## 4. 그 키워드 바탕으로 키워드 기준으로 애들 분류시키기 

```python
df['Keywords'] = ''
```

```python
df
```

```python
df['Keywords'] = '기타생활용품'

# 우선순위에 따라 특정 키워드에 대한 처리
df.loc[df['제품특성'].str.contains('체지방|다이어트|식품|성인남녀|음용|\d+정|\d+포|곤약밥|잡곡밥|초코|단백질|맛|알레르기'), 'Keywords'] = '건강식품'
df.loc[df['제품특성'].str.contains('피부'), 'Keywords'] = '화장품'
df.loc[df['제품특성'].str.contains('분유|신생아|아기밥|아기옷|아가옷|이유식|키즈|유아|어린이'), 'Keywords'] = '아기용품'
df.loc[df['제품특성'].str.contains('모발|헤어|머릿결'), 'Keywords'] = '헤어제품'

print(df)
```
